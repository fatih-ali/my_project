```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
columns = ['Sepal length','Sepal width','Petal length','Petal width','Class labels']
df = pd.read_csv('iris.csv', names=columns, skiprows=1)
display(df.head())
```

```python
df.describe()
```

```python
sns.pairplot(df,hue='Class labels')
```

```python
!wget https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv
```

```python
data = df.values

X = data[:,0:4]
Y = data[:,4]
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(Y_test.shape)
```

Support vector machine algorithm

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

Logostic regression

```python
predictions = model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictions)*100)
for i in range(len(predictions)):
  print(predictions[i], Y_test[i])
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, Y_train)
```

```python
prediction1 = model_lr.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1)*100)
for i in range(len(predictions)):
  print(predictions[i], Y_test[i])
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train, Y_train)
```

```python
prediction2 = model_dt.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)
```

```python
X_new = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])
prediction3 = model_dt.predict(X_new)
print("Prediction: {}".format(prediction3))
```
