```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("housing.csv")
display(data.head())
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
display(data.head())
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'],axis = 1)
y = data['median_house_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,10))
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(train_data.drop('ocean_proximity', axis=1).corr(),annot=True,cmap='YlOrRd')
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,10))
```

```python
train_data = train_data.join(pd.get_dummies(train_data['ocean_proximity'], dtype=int)).drop('ocean_proximity', axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(train_data.corr(),annot=True,cmap='PuRd')
```

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x = "longitude", y = "latitude", hue = "median_house_value", palette = "coolwarm", data = train_data)
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap='PuRd')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train,y_train = train_data.drop('median_house_value', axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s,y_train)
```

```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)

test_data = test_data.join(pd.get_dummies(test_data['ocean_proximity'], dtype=int)).drop('ocean_proximity', axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms']/test_data['households']
```

```python
X_test,y_test = test_data.drop('median_house_value', axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s,y_train)
```

```python
forest.score(X_test_s,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [3,10,30],
    "max_features": [2,4,6,8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(X_train_s,y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_s,y_test)
```
